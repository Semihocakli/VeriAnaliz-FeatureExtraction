#### Line Charts
- Import graph_objs as go
- Creating traces
- x = x axis
- y = y axis
- mode = type of plot like marker, line or line + markers
- name = name of the plots
- marker = marker is used with dictionary.
 - color = color of lines. It takes RGB (red, green, blue) and opacity (alpha)
- text = The hover text (hover is curser)
- data = is a list that we add traces into it
- layout = it is dictionary.
- title = title of layout
- x axis = it is dictionary
 - title = label of x axis
 - ticklen = length of x axis ticks
 - zeroline = showing zero line or not
- fig = it includes data and layout
- iplot() = plots the figure(fig) that is created by data and layout

```python
 # Line Plot

import plotly.express as px
import plotly.graph_objs as go

# Veri çerçevesini hazırla
# timesData veri çerçevesinden ilk 100 satırı seç
df = timesData.iloc[:100, :]

# Trace1 oluştur
# "Citations" için çizgi ile birbirine bağlı bir scatter plot oluştur
trace1 = go.Scatter(
    x=df.world_rank,
    y=df.citations,
    mode="lines",
    name="Alıntılar",
    marker=dict(color='rgba(16, 112, 2, 0.8)'),
    text=df.university_name
)

# Trace2 oluştur
# "Teaching" için çizgi ve işaretçili bir scatter plot oluştur
trace2 = go.Scatter(
    x=df.world_rank,
    y=df.teaching,
    mode="lines+markers",
    name="Öğretme",
    marker=dict(color='rgba(80, 26, 80, 0.8)'),
    text=df.university_name
)

# Plotly Express kullanarak figürü oluştur
# Plotly Express kullanarak, belirtilen x, y ve metin sütunlarını içeren bir scatter plot oluştur
# Daha iyi görselleştirme için özel bir renk belirle
fig = px.scatter(df, x="world_rank", y="citations", text="university_name", color_discrete_sequence=['rgba(16, 112, 2, 0.8)'],
                 title="En İyi 100 Üniversitenin Dünya Sıralamasına Göre Alıntı ve Öğretme Puanları")

# Trace2'yi figüre ekle
fig.add_trace(trace2)

# Figürü göster
fig.show()
```

#### Scatter
- Import graph_objs as go
- Creating traces
- x = x axis
- y = y axis
- mode = type of plot like marker, line or line + markers
- name = name of the plots
- marker = marker is used with dictionary.
 - color = color of lines. It takes RGB (red, green, blue) and opacity (alpha)
- text = The hover text (hover is curser)
- data = is a list that we add traces into it
- layout = it is dictionary.
- title = title of layout
- x axis = it is dictionary
 - title = label of x axis
 - ticklen = length of x axis ticks
 - zeroline = showing zero line or not
- y axis = it is dictionary and same with x axis
- fig = it includes data and layout
- iplot() = plots the figure(fig) that is created by data and layout

```python
# Scatter Plot
# Veri çerçevelerini hazırla
# timesData'dan sadece 2014, 2015 ve 2016 yıllarına ait ilk 100 üniversiteyi seç
df2014 = timesData[timesData.year == 2014].iloc[:100, :]
df2015 = timesData[timesData.year == 2015].iloc[:100, :]
df2016 = timesData[timesData.year == 2016].iloc[:100, :]

import plotly.graph_objs as go
from plotly.subplots import make_subplots

# Trace1 oluştur
trace1 = go.Scatter(
    x=df2014.world_rank,
    y=df2014.citations,
    mode="markers",
    name="2014",
    marker=dict(color='rgba(255, 128, 255, 0.8)'),
    text=df2014.university_name
)

# Trace2 oluştur
trace2 = go.Scatter(
    x=df2015.world_rank,                       # x ekseni verisi, dünya sıralaması
    y=df2015.citations,                        # y ekseni verisi, alıntılar
    mode="markers",                            # scatter plot kullanıldığı belirtiliyor
    name="2015",                               # izin adı
    marker=dict(color='rgba(255, 128, 2, 0.8)'), # marker rengi ve opaklığı
    text=df2015.university_name                # her nokta için gösterilecek metin
)

# Trace3 oluştur
trace3 = go.Scatter(
    x=df2016.world_rank,
    y=df2016.citations,
    mode="markers",
    name="2016",
    marker=dict(color='rgba(0, 255, 200, 0.8)'),
    text=df2016.university_name
)

# Verileri birleştir
data = [trace1, trace2, trace3]

# Grafik düzenini belirle
layout = go.Layout(
    title='En İyi 100 Üniversitenin 2014, 2015 ve 2016 Yıllarına Göre Alıntıları',
    xaxis=dict(title='Dünya Sıralaması', ticklen=5, zeroline=False),
    yaxis=dict(title='Alıntılar', ticklen=5, zeroline=False)
)

# Subplots oluştur
fig = make_subplots(rows=1, cols=1)

# Traces'leri figür üzerine ekleyin
fig.add_trace(trace1)
fig.add_trace(trace2)
fig.add_trace(trace3)

# Grafik düzenini güncelleyin
fig.update_layout(layout)

# Grafik gösterimini yapın
fig.show()
```

#### Bar Charts
- Import graph_objs as go
- Creating traces
- x = x axis
- y = y axis
- mode = type of plot like marker, line or line + markers
- name = name of the plots
- marker = marker is used with dictionary.
 - color = color of lines. It takes RGB (red, green, blue) and opacity (alpha)
 - line = It is dictionary. line between bars
- color = line color around bars
- text = The hover text (hover is curser)
- data = is a list that we add traces into it
- layout = it is dictionary.
- barmode = bar mode of bars like grouped
- fig = it includes data and layout
- iplot() = plots the figure(fig) that is created by data and layout

```python
# Bar Plot

# prepare data frames
df2014 = timesData[timesData.year == 2014].iloc[:3,:]

# import graph objects as "go"
import plotly.graph_objs as go

# Trace1 oluştur
trace1 = go.Bar(
    x = df2014.university_name,  # x ekseni verisi, üniversite isimleri
    y = df2014.citations,        # y ekseni verisi, alıntılar
    name = "citations",          # izin adı
    marker = dict(
        color = 'rgba(255, 174, 255, 0.5)',  # bar rengi ve opaklığı
        line=dict(color='rgb(0,0,0)',width=1.5)),  # çubuk kenar çizgisi rengi ve kalınlığı
    text = df2014.country         # her bar için gösterilecek metin
)

# create trace2
trace2 = go.Bar(
                x = df2014.university_name,
                y = df2014.teaching,
                name = "teaching",
                marker = dict(color = 'rgba(255, 255, 128, 0.5)',
                              line=dict(color='rgb(0,0,0)',width=1.5)),
                text = df2014.country)

data = [trace1, trace2]

layout = go.Layout(barmode = "group")

fig = go.Figure(data = data, layout = layout)
fig.show()
```

#### Pie Charts
- fig: create figures
- data: plot type
 - values: values of plot
 - labels: labels of plot
 - name: name of plots
 - hoverinfo: information in hover
 - hole: hole width
 - type: plot type like pie
- layout: layout of plot
 - title: title of layout
 - annotations: font, showarrow, text, x, y

```python
# data preparation
from plotly.offline import iplot

df2016 = timesData[timesData.year == 2016].iloc[:7, :]
pie1 = df2016.num_students
pie1_list = [float(each.replace(',', '.')) for each in df2016.num_students]
#num_students" sütunundaki her bir öğrenci sayısını alır ve virgülü noktaya çevirerek liste olarak pie1_list değişkenine ekler.
labels = df2016.university_name

# Figure (Pasta Grafiği)
fig = {
  "data": [
    {
      "values": pie1_list,                # Pasta dilimlerinin değerleri
      "labels": labels,                   # Pasta dilimlerinin etiketleri
      "domain": {"x": [0, .5]},           # Pasta grafiğinin alanını belirler
      "name": "Number Of Students Rates", # Pasta grafiğinin adı
      "hoverinfo":"label+percent+name",   # Hover (üzerine gelindiğinde görülecek bilgiler)
      "hole": .3,                         # Ortadaki deliğin boyutu
      "type": "pie"                       # Grafik tipi: "pie" (pasta)
    },
  ],
  "layout": {
    "title":"Üniversitelerin Öğrenci Sayıları",
    "annotations": [
      {
        "font": {"size": 20},
        "showarrow": False,
        "text": "Öğrenci Sayısı",
        "x": 0.20,
        "y": 1
      },
    ]
  }
}


# Grafiği görüntüle
iplot(fig)
```

#### Bubble Charts
- x = x axis
- y = y axis
- mode = markers(scatter)
- marker = marker properties
- color = third dimension of plot. Internaltional score
- size = fourth dimension of plot. Number of students
- text: university names

```python
# data preparation
df2016 = timesData[timesData.year == 2016].iloc[:20,:]
num_students_size  = [float(each.replace(',', '.')) for each in df2016.num_students]
international_color = [float(each) for each in df2016.international]
data = [
    {
        'y': df2016.teaching,
        'x': df2016.world_rank,
        'mode': 'markers',
        'marker': {
            'color': international_color,
            'size': num_students_size,
            'showscale': True
        },
        "text" :  df2016.university_name
    }
]
iplot(data)
```

#### Histogram
- trace1 = first histogram
- x = x axis
- y = y axis
- opacity = opacity of histogram
- name = name of legend
- marker = color of histogram
- trace2 = second histogram
- layout = layout
- barmode = mode of histogram like overlay. Also you can change it with stack

```python
# prepare data
x2011 = timesData.student_staff_ratio[timesData.year == 2011]
x2012 = timesData.student_staff_ratio[timesData.year == 2012]

trace1 = go.Histogram(
    x=x2011,
    opacity=0.75,
    name = "2011",
    marker=dict(color='rgba(171, 50, 96, 0.6)'))
trace2 = go.Histogram(
    x=x2012,
    opacity=0.75,
    name = "2012",
    marker=dict(color='rgba(12, 50, 196, 0.6)'))

data = [trace1, trace2]
layout = go.Layout(barmode='overlay',
                   title=' students-staff ratio in 2011 and 2012',
                   xaxis=dict(title='students-staff ratio'),
                   yaxis=dict( title='Count'),
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

#### Box Plot
Box Plots

- Median (50th percentile) = middle value of the data set. Sort and take the data in the middle. It is also called 50% percentile that is 50% of data are less that median(50th quartile)(quartile)

 - 25th percentile = quartile 1 (Q1) that is lower quartile
 - 75th percentile = quartile 3 (Q3) that is higher quartile
 - height of box = IQR = interquartile range = Q3-Q1
 - Whiskers = 1.5 * IQR from the Q1 and Q3
 - Outliers = being more than 1.5*IQR away from median commonly.

-trace = box
 - y = data we want to visualize with box plot
 - marker = color

```python
# data preparation
x2015 = timesData[timesData.year == 2015]

trace0 = go.Box(
    y=x2015.total_score,
    name = 'total score of universities in 2015',
    marker = dict(
        color = 'rgb(12, 12, 140)',
    )
)
trace1 = go.Box(
    y=x2015.research,
    name = 'research of universities in 2015',
    marker = dict(
        color = 'rgb(12, 128, 128)',
    )
)
data = [trace0, trace1]
iplot(data)
```

#### Scatter Matrix Plots
- import figure factory as ff
- create_scatterplotmatrix = creates scatter plot
- data2015 = prepared data. It includes research, international and total scores with index from 1 to 401
- colormap = color map of scatter plot
- colormap_type = color type of scatter plot
- height and weight

```python
# figure factory modülünü içe aktar
import plotly.figure_factory as ff

# Veriyi hazırla
dataframe = timesData[timesData.year == 2015]
data2015 = dataframe.loc[:, ["research", "international", "total_score"]]
data2015["index"] = np.arange(1, len(data2015) + 1)

# Scatter matrix (dağılım matrisi) oluştur
fig = ff.create_scatterplotmatrix(
    data2015,              # Kullanılacak veri çerçevesi
    diag='box',            # Diagonal elemanlara kutu grafiği ekle (box plot)
    index='index',         # Veri çerçevesindeki sıralama indeksi
    colormap='Portland',    # Renk haritası (colormap)
    colormap_type='cat',    # Renk haritası tipi (kategorik)
    height=700,            # Grafiğin yüksekliği
    width=700              # Grafiğin genişliği
)

# Grafiği görüntüle
iplot(fig)
```

#### Inset Plot


```python
# İlk çizgi grafiği (trace1) oluştur
trace1 = go.Scatter(
    x=dataframe.world_rank,   # x ekseninde dünya sıralamaları
    y=dataframe.teaching,     # y ekseninde öğretme puanları
    name="teaching",          # Çizgi adı
    marker=dict(color='rgba(16, 112, 2, 0.8)'),  # Çizgi rengi ve opaklığı
)

# İkinci çizgi grafiği (trace2) oluştur
trace2 = go.Scatter(
    x=dataframe.world_rank,   # x ekseninde dünya sıralamaları
    y=dataframe.income,       # y ekseninde gelir puanları
    xaxis='x2',               # x2 ekseni kullanılarak ikinci çizgiyi ayarla
    yaxis='y2',               # y2 ekseni kullanılarak ikinci çizgiyi ayarla
    name="income",            # Çizgi adı
    marker=dict(color='rgba(160, 112, 20, 0.8)'),  # Çizgi rengi ve opaklığı
)

# Verileri birleştir
data = [trace1, trace2]

# Grafik düzenini belirle
layout = go.Layout(
    xaxis2=dict(
        domain=[0.6, 0.95],     # İkinci çizgi grafiğinin x ekseni için domain belirle
        anchor='y2',            # İkinci çizgi grafiğinin y ekseni için anchor belirle
    ),
    yaxis2=dict(
        domain=[0.6, 0.95],     # İkinci çizgi grafiğinin y ekseni için domain belirle
        anchor='x2',            # İkinci çizgi grafiğinin x ekseni için anchor belirle
    ),
    title='Income and Teaching vs World Rank of Universities'  # Grafik başlığı
)

# Figürü oluştur
fig = go.Figure(data=data, layout=layout)

# Grafiği görüntüle
iplot(fig)
```

#### 3D Scatter Plot with Colorscaling
- go.Scatter3d: create 3d scatter plot
- x,y,z: axis of plots
- mode: market that is scatter
- size: marker size
- color: axis of colorscale
- colorscale: actually it is 4th dimension

```python
# 3D Scatter çizimi için trace1 oluştur
trace1 = go.Scatter3d(
    x=dataframe.world_rank,   # x ekseninde dünya sıralamaları
    y=dataframe.research,     # y ekseninde araştırma puanları
    z=dataframe.citations,    # z ekseninde atıf puanları
    mode='markers',           # Nokta tipi çizimi (scatter)
    marker=dict(
        size=10,               # Nokta boyutu
        color='rgb(255,0,0)',  # Nokta rengi (RGB formatında)
    )
)

# Verileri birleştir
data = [trace1]

# Grafik düzenini belirle
layout = go.Layout(
    margin=dict(
        l=0,  # Sol kenar boşluğu
        r=0,  # Sağ kenar boşluğu
        b=0,  # Alt kenar boşluğu
        t=0   # Üst kenar boşluğu
    )
)

# Figürü oluştur
fig = go.Figure(data=data, layout=layout)

# Grafiği görüntüle
iplot(fig)
```

#### Multiple Subplots


```python
trace1 = go.Scatter(
    x=dataframe.world_rank,
    y=dataframe.research,
    name = "research"
)
trace2 = go.Scatter(
    x=dataframe.world_rank,
    y=dataframe.citations,
    xaxis='x2',
    yaxis='y2',
    name = "citations"
)
trace3 = go.Scatter(
    x=dataframe.world_rank,
    y=dataframe.income,
    xaxis='x3',
    yaxis='y3',
    name = "income"
)
trace4 = go.Scatter(
    x=dataframe.world_rank,
    y=dataframe.total_score,
    xaxis='x4',
    yaxis='y4',
    name = "total_score"
)
data = [trace1, trace2, trace3, trace4]
layout = go.Layout(
    xaxis=dict(
        domain=[0, 0.45]
    ),
    yaxis=dict(
        domain=[0, 0.45]
    ),
    xaxis2=dict(
        domain=[0.55, 1]
    ),
    xaxis3=dict(
        domain=[0, 0.45],
        anchor='y3'
    ),
    xaxis4=dict(
        domain=[0.55, 1],
        anchor='y4'
    ),
    yaxis2=dict(
        domain=[0, 0.45],
        anchor='x2'
    ),
    yaxis3=dict(
        domain=[0.55, 1]
    ),
    yaxis4=dict(
        domain=[0.55, 1],
        anchor='x4'
    ),
    title = 'Research, citation, income and total score VS World Rank of Universities'
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
